## Generate Graph

```python
A = np.zeros((50,50))
for i in range(25):
    for j in range(25):
        A[i,j] = 1; A[j,i] = 1;
        A[i+25,j+25] = 1; A[j+25,i+25] = 1
        
A[0,25] = 1; A[25,0] = 1
```

```python
G = graphs.Graph(A)
G.set_coordinates()
```

# Call The Sparsifier

```python
sparsifier = sparsify.Sparsifier(G, epsilon = 0.2)
```

```python
sparsifier.sparsify(verbose=True)
G_ = sparsifier.get_sparsifier()
```
